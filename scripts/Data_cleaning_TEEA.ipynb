```python
import pandas as pd
```

```python
df_id = pd.read_csv("id_table.csv")
df_id
```

```python
df_id.set_index('file_name',inplace=True)
df_id
```

```python
my_dict = df_id.to_dict()['barcode']
my_dict
```

```python
uuids = pd.read_csv("uuids.csv")
uuids
```

```python
import numpy as np
for column in uuids:
    columnSeriesObj = uuids[column]
    #print(columnSeriesObj.values)
    print(type(columnSeriesObj.values))
    a = columnSeriesObj.values
    a = a.tolist()
    print(type(a))
    print(a)
```

```python
answer = []
for id in a:
    answer.append(my_dict[id])
    print(answer)
```

```python
answer
```

```python
df = pd.DataFrame(answer)
df
```
